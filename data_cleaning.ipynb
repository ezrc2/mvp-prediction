```python
import pandas as pd
```

### Cleaning MVP data

```python
mvp_df = pd.read_csv('mvps.csv')
mvp_df.head()
```

```python
# Remove duplicate stats from players.csv
mvp_df = mvp_df[['Player', 'Year', 'Share']]
mvp_df.head()
```

### Cleaning player data

```python
player_df = pd.read_csv('players.csv')
player_df.head()
```

```python
player_df.drop('Rk', axis=1, inplace=True)
player_df.head()
```

```python
# Removing * after some names
player_df['Player'] = player_df['Player'].str.replace('*', '', regex=False)
player_df.head()
```

```python
# consolidate rows of players who played for multiple seasons that year into one row
def consolidate_rows(df):
    if df.shape[0] == 1:
        return df
    row = df[df['Tm'] == 'TOT'] # total
    row['Tm'] = df['Tm'].iloc[-1]
    return row
```

```python
player_df = player_df.groupby(['Player', 'Year']).apply(consolidate_rows)
```

```python
player_df.head()
```

```python
player_df.index = player_df.index.droplevel()
player_df.index = player_df.index.droplevel()
player_df.head()
```

#### Combining MVP and player data

```python
combined = player_df.merge(mvp_df, how='outer', on=['Player', 'Year'])
combined.head()
```

```python
combined[['Share']] = combined[['Share']].fillna(0)
combined.head()
```

```python
combined[combined['Share'] > 0]
```

#### Team records

```python
record_df = pd.read_csv('records.csv')
record_df.head(20)
```

```python
# Remove rows with the division
record_df = record_df[~record_df['Team'].str.contains('Division')]
record_df.head(20)
```

```python
# remove * from team names
record_df['Team'] = record_df['Team'].str.replace('*', '', regex=False)
record_df.head()
```

```python
stats_df = combined.merge(record_df, how='outer', on=['Team', 'Year'])
stats_df.head()
```
